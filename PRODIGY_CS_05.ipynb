```python
import random
import time
import json

# Disclaimer
print("------------------------ Packet Sniffer Tool Disclaimer ---------------------------")
print("This packet sniffer tool is intended for educational and ethical purposes only.")
print("By using this tool, you agree to use it only on networks for which you have explicit permission.")
print("Unauthorized use of this tool is strictly prohibited.")

# Function to simulate packet generation
def generate_packet():
    protocols = ['TCP', 'UDP', 'ICMP', 'HTTP', 'HTTPS']
    src_ip = f"192.168.1.{random.randint(1, 255)}"
    dst_ip = f"192.168.1.{random.randint(1, 255)}"
    protocol = random.choice(protocols)
    payload = "Sample payload data " + str(random.randint(1000, 9999))

    return {
        "source_ip": src_ip,
        "destination_ip": dst_ip,
        "protocol": protocol,
        "payload": payload
    }

# Simulated packet capturing
def packet_sniffer(duration=10):
    end_time = time.time() + duration
    captured_packets = []

    print("\n--------------- Packet Sniffing Simulation ---------------")
    while time.time() < end_time:
        packet = generate_packet()
        captured_packets.append(packet)

        # Display captured packet information
        print(json.dumps(packet, indent=4))
        time.sleep(1)  # Simulate delay between packet captures

    print("\n---------------------- Simulation Complete ----------------------")
    print(f"Total packets captured: {len(captured_packets)}")

# Start the simulated packet sniffer for 10 seconds
packet_sniffer(duration=10)
```
